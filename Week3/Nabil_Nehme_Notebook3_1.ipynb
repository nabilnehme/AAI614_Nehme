## Random Walk

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import datetime
import random
import re

count=0
random.seed(datetime.datetime.now().strftime('%s'))
def getLinks(articleUrl):
    html = urlopen(f'http://en.wikipedia.org{articleUrl}')
    bs = BeautifulSoup(html, 'html.parser')
    return bs.find('div', {'id':'bodyContent'}).find_all('a', href=re.compile('^(/wiki/)((?!:).)*$'))

links = getLinks('/wiki/Kevin_Bacon')
while len(links) > 0 and count < 10:
    newArticle = links[random.randint(0, len(links)-1)].attrs['href']
    print(newArticle)
    links = getLinks(newArticle)
    count+=1
```

## Recursively crawling an entire site

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

pages = set()
max_links = 100  # Set the maximum number of links to process
link_count = 0  # Global variable to track link count

def getLinks(pageUrl):
    global link_count  # Access the global count variable

    if link_count >= max_links:
        return

    # Check if pageUrl starts with '/wiki/' to form valid URL
    if not pageUrl.startswith('/wiki/'):
        fullUrl = f'http://en.wikipedia.org/wiki/{pageUrl}'
    else:
        fullUrl = f'http://en.wikipedia.org{pageUrl}'

    # Handle HTTP errors, such as 404
    try:
        html = urlopen(fullUrl)
    except Exception as e:
        print(f"Error opening {fullUrl}: {e}")
        return

    bs = BeautifulSoup(html, 'html.parser')
    for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
        if 'href' in link.attrs:
            if link.attrs['href'] not in pages:
                # We have encountered a new page
                newPage = link.attrs['href']
                print(newPage)
                pages.add(newPage)
                link_count += 1  # Increment the global count
                getLinks(newPage)

getLinks('Kevin_Bacon') # Start with a specific page instead of ''
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

pages = set()
max_links = 100  # Set the maximum number of links to process
link_count = 0  # Global variable to track link count

def getLinks(pageUrl):
    global link_count  # Access the global count variable

    if link_count >= max_links:
        return

    # Check if pageUrl is empty, if so, assume it's the starting page
    if not pageUrl:
        pageUrl = '/wiki/Kevin_Bacon'  # Start with a specific page instead of ''

    fullUrl = f'http://en.wikipedia.org{pageUrl}'
    print(f"Trying to open: {fullUrl}")  # Print the URL for debugging

    try:
        html = urlopen(fullUrl)
    except Exception as e:
        print(f"Error opening {fullUrl}: {e}")
        return

    bs = BeautifulSoup(html, 'html.parser')
    for link in bs.find_all('a', href=re.compile('^(/wiki/)((?!:).)*$')):
        if 'href' in link.attrs:
            newPage = link.attrs['href']

            # Check if newPage is already visited or is an internal link
            if newPage not in pages and ':' not in newPage:
                # Build a valid URL by joining the base URL with newPage
                fullNewPageUrl = f"http://en.wikipedia.org{newPage}"
                print(f"Found new page: {fullNewPageUrl}")
                pages.add(newPage)
                link_count += 1  # Increment the global count
                getLinks(newPage)  # Recursively call getLinks for the new page

getLinks('') # Now this will start with '/wiki/Kevin_Bacon'
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

pages = set()
max_links = 100  # Set the maximum number of links to process
link_count = 0  # Global variable to track link count

def getLinks(pageUrl):
    global link_count  # Access the global count variable

    # Check if the maximum link count has been reached
    if link_count >= max_links:
        print(f"Reached maximum link count of {max_links}. Stopping.")
        return

    fullUrl = f'http://en.wikipedia.org{pageUrl}'

    print(f"Trying to open: {fullUrl}")  # Print the URL for debugging

    try:
        html = urlopen(fullUrl)
    except Exception as e:
        print(f"Error opening {fullUrl}: {e}")
        return  # Stop processing if there's an error

    bs = BeautifulSoup(html, 'html.parser')
    try:
        print(bs.h1.get_text())
        bodyContent = bs.find('div', {'id':'bodyContent'}).find_all('p')
        if len(bodyContent):
            print(bodyContent[0])
        print(bs.find(id='ca-edit').find('a').attrs['href'])
    except AttributeError:
        print('This page is missing something! Continuing.')

    for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
        if 'href' in link.attrs:
            newPage = link.attrs['href']

            # Check if newPage is already visited or is an internal link
            if newPage not in pages and ':' not in newPage:
                # Print the new page found to track the process
                print('-'*20)
                print(f"Found new page: {newPage}")

                pages.add(newPage)
                link_count += 1  # Increment the global count

                # Check if the maximum link count has been reached after incrementing
                if link_count >= max_links:
                    print(f"Reached maximum link count of {max_links}. Stopping.")
                    return

                getLinks(newPage)  # Recursively call getLinks for the new page

# This is called with a valid URL, likely the 404 error occurs after several redirect.
getLinks('/wiki/General-purpose_programming_language')
```

## Crawling across the Internet

```python
from urllib.request import urlopen
from urllib.parse import urlparse
from bs4 import BeautifulSoup
import re
import datetime
import random

#Retrieves a list of all Internal links found on a page
def getInternalLinks(bs, url):
    netloc = urlparse(url).netloc
    scheme = urlparse(url).scheme
    internalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc == '':
            internalLinks.add(f'{scheme}://{netloc}/{link.attrs["href"].strip("/")}')
        elif parsed.netloc == netloc:
            internalLinks.add(link.attrs['href'])
    return list(internalLinks)

#Retrieves a list of all external links found on a page
def getExternalLinks(bs, url):
    netloc = urlparse(url).netloc
    externalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc != '' and parsed.netloc != netloc:
            externalLinks.add(link.attrs['href'])
    return list(externalLinks)

def getRandomExternalLink(startingPage):
    bs = BeautifulSoup(urlopen(startingPage), 'html.parser')
    externalLinks = getExternalLinks(bs, startingPage)
    if not len(externalLinks):
        print('No external links, looking around the site for one')
        internalLinks = getInternalLinks(bs, startingPage)
        return getRandomExternalLink(random.choice(internalLinks))
    else:
        return random.choice(externalLinks)

def followExternalOnly(startingSite):
    externalLink = getRandomExternalLink(startingSite)
    print(f'Random external link is: {externalLink}')
    followExternalOnly(externalLink)


followExternalOnly('https://www.oreilly.com/')
```

```python
from urllib.request import urlopen
from urllib.parse import urlparse
from bs4 import BeautifulSoup
import re
import datetime
import random
import time

#Retrieves a list of all Internal links found on a page
def getInternalLinks(bs, url):
    netloc = urlparse(url).netloc
    scheme = urlparse(url).scheme
    internalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        # Check if the link is relative
        if parsed.netloc == '' and parsed.scheme == '':
            # Construct a full URL using the starting page's scheme and netloc
            internalLinks.add(f'{scheme}://{netloc}/{link.attrs["href"].strip("/")}')
        elif parsed.netloc == netloc:
            internalLinks.add(link.attrs['href'])
    return list(internalLinks)

#Retrieves a list of all external links found on a page
def getExternalLinks(bs, url):
    netloc = urlparse(url).netloc
    externalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        # Check for valid external links
        if parsed.netloc != '' and parsed.netloc != netloc and parsed.scheme in ('http', 'https'):
            externalLinks.add(link.attrs['href'])
    return list(externalLinks)

def getRandomExternalLink(startingPage):
    # Added a delay to avoid overwhelming the server
    time.sleep(random.uniform(1, 3))

    try:
        bs = BeautifulSoup(urlopen(startingPage), 'html.parser')
    except Exception as e:
        print(f"Error opening {startingPage}: {e}")
        return None  # Return None to indicate an error

    externalLinks = getExternalLinks(bs, startingPage)
    if not len(externalLinks):
        print('No external links, looking around the site for one')
        internalLinks = getInternalLinks(bs, startingPage)
        if internalLinks:  # Check if internalLinks is not empty
            return getRandomExternalLink(random.choice(internalLinks))
        else:
            return None  # Return None if no internal links found either
    else:
        return random.choice(externalLinks)

def followExternalOnly(startingSite, max_iterations=5):  # Added a max_iterations parameter
    """Follows external links for a given number of iterations."""

    for _ in range(max_iterations):  # Iterate for a maximum number of times
        externalLink = getRandomExternalLink(startingSite)
        if externalLink is None:  # Handle the case where getRandomExternalLink returns None
            print("Could not find an external link to follow.")
            break  # Exit the loop if no external link is found

        print(f'Random external link is: {externalLink}')
        startingSite = externalLink  # Update startingSite for the next iteration

    print("Finished following external links.")

followExternalOnly('https://www.oreilly.com/')
```

## Collect all External Links from a Site

```python
# Collects a list of all external URLs found on the site
allExtLinks = []
allIntLinks = []


def getAllExternalLinks(url):
    bs = BeautifulSoup(urlopen(url), 'html.parser')
    internalLinks = getInternalLinks(bs, url)
    externalLinks = getExternalLinks(bs, url)
    for link in externalLinks:
        if link not in allExtLinks:
            allExtLinks.append(link)
            print(link)

    for link in internalLinks:
        if link not in allIntLinks:
            allIntLinks.append(link)
            getAllExternalLinks(link)


allIntLinks.append('https://oreilly.com')
getAllExternalLinks('https://www.oreilly.com/')
```
