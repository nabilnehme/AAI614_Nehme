```python
# Create a new column, pre_tip_amount
df['pre_tip_amount'] = (
    df
    [['fare_amount', 'extra', 'mta_tax', 'tolls_amount',
      'improvement_surcharge', 'congestion_surcharge']]
    .sum(axis='columns')
)
```

```python
# Create a new column, tip_percentage, showing the percent of the fare_amount that people tipped
df['tip_percentage'] = df['tip_amount'] / df['pre_tip_amount']
```

```python
# What was the overall tip percentage?
df['tip_percentage'].mean()
```

```python
# How many times did people tip more than 100%?
(df['tip_percentage'] > 1).value_counts()
```

```python
# What percent of taxi riders give no tip at all?
(df['tip_percentage'] == 0).value_counts(normalize=True)
```

# Questions

1. We saw that 32 percent of riders don't tip at all. Of those who *do*, what percentage do they tip, on average?

```python
# Filter the dataset to include only rows where tip_percentage is greater than 0 (those who tipped)
df_tippers = df[df['tip_percentage'] > 0]

# Calculate the average tip percentage for those who tipped
average_tip_for_tippers = df_tippers['tip_percentage'].mean()

average_tip_for_tippers
```

2. How many of the rides in our data set, supposedly from January and July 2019, are from outside of those dates?

```python
# Filter the dataset for rides outside of January and July 2019
df_outside_jan_jul = df[~((df['tpep_pickup_datetime'] >= '2019-01-01') & (df['tpep_pickup_datetime'] <= '2019-01-31')) &
                        ~((df['tpep_pickup_datetime'] >= '2019-07-01') & (df['tpep_pickup_datetime'] <= '2019-07-31'))]

# Count the number of rows outside these dates
num_outside_jan_jul = df_outside_jan_jul.shape[0]

num_outside_jan_jul
```

3. Looking only at dates in January and July, on what week did passengers tip the greatest percentage?

```python
# Ensure 'tpep_pickup_datetime' is in datetime format
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'], errors='coerce')

# Filter only for January and July dates, and keep only relevant columns
df_filtered = df[
    ((df['tpep_pickup_datetime'] >= '2019-01-01') & (df['tpep_pickup_datetime'] <= '2019-01-31')) |
    ((df['tpep_pickup_datetime'] >= '2019-07-01') & (df['tpep_pickup_datetime'] <= '2019-07-31'))
][['tpep_pickup_datetime', 'tip_percentage']]

# Set the 'tpep_pickup_datetime' column as the index for resampling
df_filtered = df_filtered.set_index('tpep_pickup_datetime')

# Resample by week and calculate the mean 'tip_percentage' for each week
week_tip_percentage = df_filtered['tip_percentage'].resample('W').mean()

# Sort the weeks by the average 'tip_percentage' in descending order and get the top result
week_tip_percentage_sorted = week_tip_percentage.sort_values(ascending=False)

# Get the week with the greatest average tip percentage
week_tip_percentage_sorted.head(1)



```
